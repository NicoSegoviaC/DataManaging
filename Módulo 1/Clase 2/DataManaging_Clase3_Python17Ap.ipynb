```python
##Ejemplo 1
# Importar librerías
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar datos
iris = sns.load_dataset("iris")


iris.head(20)


Dropbox/CLASES/DataManaging/Clase3/data/mtcars.csv
```

```python
# Calcular promedio y desviación estándar de la longitud del pétalo por especie
iris_summary = iris.groupby("species").agg(
    sepal_length_mean=("sepal_length", "mean"),
    sepal_length_sd=("sepal_length", "std"),
    sepal_width_mean=("sepal_width", "mean"),
    sepal_width_sd=("sepal_width", "std"),
    petal_length_mean=("petal_length", "mean"),
    petal_length_sd=("petal_length", "std"),
    petal_width_mean=("petal_width", "mean"),
    petal_width_sd=("petal_width", "std")
)

print(iris_summary)
```

```python
# Gráfico de dispersión con color por especie y etiquetas en los ejes
sns.scatterplot(data=iris, x="sepal_length", y="petal_length", hue="species")
plt.xlabel("Longitud del Sépalo (cm)")
plt.ylabel("Longitud del Pétalo (cm)")
plt.show()
```

```python
# Importación del conjunto de datos iris

from sklearn.datasets import load_iris

# Carga del conjunto de datos iris
iris = load_iris()

# Ajuste del modelo de análisis discriminante
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
modelo = LinearDiscriminantAnalysis().fit(iris.data, iris.target)

# Obtención de las coordenadas proyectadas en ambos discriminantes lineales
coordenadas = modelo.transform(iris.data)

# Agregar la variable target a las coordenadas proyectadas
import pandas as pd
coordenadas = pd.DataFrame(coordenadas, columns=['LD1', 'LD2'])
coordenadas['target'] = iris.target

print(coordenadas)
```

```python
# Resumen del modelo de análisis discriminante
print('Prior probabilities of groups:\n', modelo.priors_)
print('\nGroup means:\n', modelo.means_)
print('\nCoefficients of linear discriminants:\n', modelo.coef_)
print('\nProportion of trace:\n', modelo.explained_variance_ratio_)
```

```python
# Gráfico de los resultados del análisis discriminante
import matplotlib.pyplot as plt
plt.scatter(modelo.transform(iris.data)[:, 0], modelo.transform(iris.data)[:, 1], c=iris.target, cmap='rainbow')
plt.title("Análisis discriminante")
plt.show()
```

```python
# Cargar datos
mtcars = pd.read_csv("mtcars.csv")

mtcars.head()
```

```python
# Resumen de los datos
mtcars.describe()

# Calcular promedio y desviación estándar de millas por galón (mpg) por número de cilindros

mtcars_summary = mtcars.groupby("cyl").agg(
    mpg_mean=("mpg", "mean"),
    mpg_sd=("mpg", "std")
).reset_index()

print(mtcars_summary)
```

```python
# Gráfico de barras de millas por galón (mpg) promedio por número de cilindros
sns.barplot(data=mtcars_summary, x="cyl", y="mpg_mean", color="steelblue")
plt.errorbar(
    x=mtcars_summary.index, y=mtcars_summary["mpg_mean"],
    yerr=mtcars_summary["mpg_sd"], fmt="none", ecolor="black", capsize=5
)
plt.xlabel("Número de cilindros")
plt.ylabel("Millas por galón (mpg) promedio")
plt.show()
```

```python
# Crear la expresión regular para extraer las dos primeras palabras
##La expresión regular r'^(\w)\w+\s(\w+)' busca al principio de la cadena (^) 
##una palabra que comience con una letra ((\w)), seguida de una o más letras (\w+), 
##un espacio en blanco (\s), y finalmente otra palabra de una o más letras ((\w+)). 
##La expresión regular crea dos grupos, uno para la primera letra de la primera palabra, 
##y otro para la segunda palabra completa.

import re
pattern = re.compile(r'^(\w)\w+\s(\w+)')

#Luego, utilizamos apply() para aplicar la expresión regular a cada valor en la columna "Scientific_name" y reemplazarlo con la letra y la segunda palabra unidos por un guión bajo.
# Modificar la columna "Scientific_name" utilizando la expresión regular

df_sp["Scientific_name"] = df_sp["Scientific_name"].apply(lambda x: pattern.sub(r'\1_\2', x))

# Guardar el archivo modificado como CSV
#df.to_csv("tu_archivo_modificado.csv", index=False)

df_sp.head()

```
